```python
#get_dummies() method is used to convert the categorical features of the data into dummy variables or indicator variables.
nn_df = pd.get_dummies(df.Name).iloc[:,0:]
print(nn_df)
```

```python
#select_dtypes() and include np.number to select numerical columns whereas include objects for categorical columns
import numpy as np
m_df = df.select_dtypes(include=[np.number])
m_df.head()
```

```python
mm_df = df.select_dtypes(include='object')
mm_df.head()
```

```python
#concatenation of pandas object into a DataFrame output along a particular axis with optional set logic such as union and intersection using concat() method
concat_df = pd.concat([m_df,mm_df],axis=1)
print(concat_df)
```

```python
#qcut() method ensures a more even distribution of the values inside each bin, so we can say it’s a better sampling.
#cut() method is used to specifically define the bin edges and hence the distribution of values is not even across all the bins
df["Speed_cut"] = pd.cut(df["Speed"],bins=3,labels=['1-30','31-50','50-70'])
```

```python
df["Speed_cut_q"] = pd.qcut(df["Speed"],q=5)
```
